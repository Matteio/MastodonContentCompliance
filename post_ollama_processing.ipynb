```python
#un thread demone avvia il server locale di ollama
import subprocess
import threading
t = threading.Thread(target=lambda: subprocess.run(["ollama", "serve"]),daemon=True)
t.start()
```

```python
#un altro thread demone avvia llama3
!ollama pull llama3
t2 = threading.Thread(target=lambda: subprocess.run(["ollama", "run", "llama3"]),daemon=True)
t2.start()
```

```python
import requests
import json
def ask_to_llama(context, prompt):
  full_prompt = context + prompt
  response = requests.post('http://localhost:11434/api/generate', data = json.dumps({'model':'llama3', 'prompt': full_prompt}))
  jsons = [json.loads(s) for s in response.text.split('\n') if s!=''] #ollama restituisce un oggetto json per ogni token, ciascuno su una riga
  s = ''
  for elem in jsons:
    s += elem['response']
  return s
```

```python
ask_to_llama('', 'What are main italian cities?')
```
