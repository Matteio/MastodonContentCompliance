```python
from google.colab import drive
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
drive.mount('/content/drive', force_remount = True)
```

```python
counters = json.load(open('/content/drive/MyDrive/[Capalbo - Gabriele] Mastodon Content Compliance/data/counters.json','r'))['instances']
```

```python
counters_sorted = sorted(counters, key = lambda x:x['counter'], reverse = True)
```

```python
yplot = [elem['counter'] for elem in counters_sorted]
logy = np.log10(yplot)
```

```python
fig, ax = plt.subplots()
ax.plot(range(1,len(yplot)+1), yplot)
ax.set_xlabel('i-th bigger instance')
ax.set_ylabel('millions of posts')
ax.set_title('instance size vs number of new posts')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(np.log10(range(1,len(yplot)+1)), logy)
ax.set_xlabel('i-th bigger instance (log)')
ax.set_ylabel('millions of posts (log)')
ax.set_title('instance size vs number of new posts in a log-log space')
plt.show()
```
