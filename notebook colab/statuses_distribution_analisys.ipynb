```python
from google.colab import drive
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
drive.mount('/content/drive', force_remount = True)
```

```python
counters = json.load(open('/content/drive/MyDrive/[Capalbo - Gabriele] Mastodon Content Compliance/data/counters.json','r'))['instances']
```

```python
counters_sorted = sorted(counters, key = lambda x:x['counter'], reverse = True)
```

```python
yplot = [elem['counter'] for elem in counters_sorted]
logy = np.log10(yplot)
```

```python
plt.plot(range(len(yplot)), yplot)
plt.show()
```

```python
plt.plot(np.log10(range(len(yplot))), logy)
plt.show()
```
